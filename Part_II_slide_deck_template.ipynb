## Investigation Overview

In this slide deck, I will be investigating trends in the prosper loan dataset. Particular focus will be on trends in the loan amount distribution, loan collection rate over the years and the term distribution amongst loanees.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

## Importing the dataset


```python
df=pd.read_csv("prosperLoanData.csv")
```

## Loan Amounts Distribution

In the first plot, I'll be investigating the distribution of loan amounts to discover the frequently occuring loan amounts.

```python
# plot a histogram showing the distribution of the Original Loan Amount
plt.hist(data = df, x = 'LoanOriginalAmount', bins = 20)
plt.xlabel('Loan Amount')
plt.ylabel('Count')
plt.title('Loan Amount Distribution');
```

## Tenor Distribution

In the next plot, I'll be using a doughnut chart to show the tenors and their respective frequency of occurrence.

```python
tenor = df['Term'].value_counts()
plt.pie(tenor, labels = tenor.index, startangle = 90, counterclock = False, wedgeprops = {'width' : 0.3})
plt.axis('square')
plt.title('Term Distribution');
```

## Borrower Rate Distribution

In the next plot, I'll examine the distribution of borrower rate for any trends and patterns

```python
# Multiplying the BorrowerRate column by 100 to get percentage figure

df['BorrowerRate'] = df['BorrowerRate']*100

plt.figure(figsize = [20,5])

plt.subplot(1,2,1)
bin_vals = np.arange(0.0, df['BorrowerRate'].max(), 1)
plt.hist(data = df, x = 'BorrowerRate', bins = bin_vals);
plt.xlabel('Borrower Rate')
plt.ylabel('Count')
plt.title('Borrower Rate Distribution')


plt.subplot(1,2,2)
bin_vals = np.arange(-0.5, df['BorrowerRate'].max()+1.5, 1)
plt.hist(data = df, x = 'BorrowerRate', bins = bin_vals, rwidth = 0.7)
plt.xlabel('Borrower Rate')
plt.ylabel('Count')
plt.title('Borrower Rate Distribution');
```

## Loan Collection Over the Years

Last but not least, I will access loan collection count by year and observe the trends across the years captured in the dataset.

```python

df['Year'] = pd.DatetimeIndex(df['ListingCreationDate']).year
sns.countplot(data=df, x='Year', color = sns.color_palette()[0])
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Loan Collection Across the Years');
```
